```python
# Handling Bulk Stage
df_bulk_days_supporter = pd.read_excel("C:/Users/DKC/Desktop/Python Server Processing Code/Excel Files/bulk.xlsx")

def get_bulk_days(process, quantity):
    # Filter the DataFrame based on the process
    process_data = df_bulk_days_supporter[df_bulk_days_supporter['process'] == process]
    
    # Iterate through the rows to find the matching quantity range
    for _, row in process_data.iterrows():
        if row['minimum_quantity'] <= int(quantity) <= row['maximum_quantity']:
            return row['minimum_days']
    
    return "Quantity out of range or process not found"

for fabrics_details in user_submitted_data["fabricDetails"]:
    fabric_name = fabrics_details["name"]
    
    if "SHIFFLY" not in fabrics_details["processes"]:
        for process in fabrics_details["processes"]:
            start_date = df_workflow_tracker[
                (df_workflow_tracker["Stage"] == "FOB") & 
                (df_workflow_tracker["Fabric Name"] == fabric_name) & 
                (df_workflow_tracker["Process"] == process)
            ]["Original End Date"].max()
            
            for color in fabrics_details["colors"]:
                lowest_days = get_bulk_days(process, color["quantity"])
                end_date = start_date + timedelta(days=lowest_days)
                df_workflow_tracker = add_to_dataframe(
                    df_workflow_tracker,
                    stage="BULK",
                    fabric_name=fabric_name,
                    process=process,
                    description=f"{process} process for {fabric_name}",
                    quantity=color["quantity"],
                    color=color["name"],
                    start_date=start_date,
                    lowest_days=lowest_days,
                    end_date=end_date
                )
    else:
        # Handle SHIFFLY and DYING processes
        if "DYING" in fabrics_details["processes"]:
            # Start SHIFFLY process
            shiffly_start_date = user_submitted_data["approvalDate"]  # Assuming approvalDate is the start date for SHIFFLY
            shiffly_days = get_bulk_days("SHIFFLY", sum(color["quantity"] for color in fabrics_details["colors"]))
            shiffly_end_date = shiffly_start_date + timedelta(days=shiffly_days)
            
            df_workflow_tracker = add_to_dataframe(
                df_workflow_tracker,
                stage="BULK",
                fabric_name=fabric_name,
                process="SHIFFLY",
                description=f"SHIFFLY process for {fabric_name}",
                quantity=sum(color["quantity"] for color in fabrics_details["colors"]),
                color="null",
                start_date=shiffly_start_date,
                lowest_days=shiffly_days,
                end_date=shiffly_end_date
            )
            
            # Start DYING process for each color after 2 days of SHIFFLY end date
            dying_start_date = shiffly_end_date + timedelta(days=2)
            for color in fabrics_details["colors"]:
                dying_days = get_bulk_days("DYING", color["quantity"])
                dying_end_date = dying_start_date + timedelta(days=dying_days)
                
                df_workflow_tracker = add_to_dataframe(
                    df_workflow_tracker,
                    stage="BULK",
                    fabric_name=fabric_name,
                    process="DYING",
                    description=f"DYING SHIFFLY for {fabric_name} - {color['name']}",
                    quantity=color["quantity"],
                    color=color["name"],
                    start_date=dying_start_date,
                    lowest_days=dying_days,
                    end_date=dying_end_date
                )

df_workflow_tracker
```

```python
import pandas as pd
from datetime import datetime, timedelta

# Example user input
user_input = {
    "order_details": {
        "order_id": "12345",
        "order_date": "2024-09-01",
        "buyer": "ABC Company",
        "delivery_date": "2024-12-01"
    },
    "fabric_details": [
        {
            "fabric_name": "Cotton",
            "quantity": 1000,
            "is_imported": False,
            "processes": [
                {
                    "name": "Dyeing",
                    "details": {
                        "colors": [
                            {"color_name": "Red", "quantity": 500},
                            {"color_name": "Blue", "quantity": 500}
                        ]
                    }
                },
                {
                    "name": "Printing",
                    "details": {
                        "patterns": ["Stripes", "Polka Dots"]
                    }
                },
                {
                    "name": "Stitching",
                    "details": {
                        "style": "T-Shirt"
                    }
                }
            ]
        }
    ],
    "constraints": {
        "deadlines": {
            "Dyeing": "2024-09-10",
            "Printing": "2024-09-20",
            "Stitching": "2024-10-01"
        },
        "skip_stages": ["Embroidery"]
    }
}

# Define stages and dependencies
stages = {
    "Procurement": {"id": 1, "duration": 7},
    "Dyeing": {"id": 2, "duration": 5},
    "Printing": {"id": 3, "duration": 7},
    "Embroidery": {"id": 4, "duration": 10},
    "Stitching": {"id": 5, "duration": 15}
}

dependencies = {
    1: [2],  # Procurement followed by Dyeing
    2: [3, 4],  # Dyeing followed by Printing and Embroidery
    3: [5],  # Printing followed by Stitching
    4: [5]  # Embroidery followed by Stitching
}

# Initialize workflow DataFrame
workflow = pd.DataFrame(columns=["fabric_name", "stage", "start_date", "end_date", "status"])

# Function to add stage to the workflow
def add_stage(fabric_name, stage, start_date, duration):
    end_date = start_date + timedelta(days=duration)
    workflow.loc[len(workflow)] = [fabric_name, stage, start_date, end_date, "Pending"]
    return end_date

# Start processing each fabric
for fabric in user_input["fabric_details"]:
    fabric_name = fabric["fabric_name"]
    order_date = datetime.strptime(user_input["order_details"]["order_date"], "%Y-%m-%d")
    current_date = order_date + timedelta(days=3)  # Start 3 days after the order date

    # Procurement stage
    current_date = add_stage(fabric_name, "Procurement", current_date, stages["Procurement"]["duration"])

    for process in fabric["processes"]:
        process_name = process["name"]
        if process_name in user_input["constraints"]["skip_stages"]:
            continue

        duration = stages[process_name]["duration"]
        deadline = user_input["constraints"]["deadlines"].get(process_name)
        if deadline:
            current_date = datetime.strptime(deadline, "%Y-%m-%d")
        current_date = add_stage(fabric_name, process_name, current_date, duration)

# Final workflow
print(workflow)
```
